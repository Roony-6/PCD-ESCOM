```python
import re
from typing import Dict, List
import datetime
import csv

# Departamentos válidos para empleados
DEPARTAMENTOS_VALIDOS = ['VEN', 'ADM', 'TEC', 'LOG', 'RHH']

# Series válidas para facturas
SERIES_VALIDAS = ['A', 'B', 'C', 'D', 'E']
```

```python
def validar_producto(codigo: str) -> Dict:
    """
    Valida código de producto.
    Formato: ABC-1234-MX
    Patrón regex: ^([A-Z]{3})-(\d{4})-([A-Z]{2})$
      - [A-Z]{3}  → exactamente 3 letras mayúsculas (categoría)
      - \d{4}     → exactamente 4 dígitos (número)
      - [A-Z]{2}  → exactamente 2 letras mayúsculas (país)
    """
    resultado = {"valido": False, "categoria": None, "numero": None, "pais": None}
    # Anclas ^ y $ garantizan coincidencia total, sin caracteres extra
    patron = r'^([A-Z]{3})-(\d{4})-([A-Z]{2})$'
    m = re.match(patron, codigo)
    if m:
        resultado["valido"]    = True
        resultado["categoria"] = m.group(1)  # ej. "TEC"
        resultado["numero"]    = m.group(2)  # ej. "0001"
        resultado["pais"]      = m.group(3)  # ej. "MX"
    return resultado
```

```python
def validar_envio(codigo: str) -> Dict:
    """
    Valida código de envío.
    Formato: ENV-YYYY-MM-DD-NNNNNN
    Patrón regex: ^ENV-(20(?:2[0-9]|30))-(\d{2})-(\d{2})-(\d{6})$
      - 20(?:2[0-9]|30) → años 2020-2030 (no se captura el grupo interno)
      - \d{2} para mes → validación de rango 01-12 por código Python
      - \d{2} para día → validación de rango 01-31 por código Python
      - \d{6}           → 6 dígitos secuenciales
    """
    resultado = {"valido": False, "fecha": None, "secuencial": None}
    patron = r'^ENV-(20(?:2[0-9]|30))-(\d{2})-(\d{2})-(\d{6})$'
    m = re.match(patron, codigo)
    if m:
        anio, mes, dia = int(m.group(1)), int(m.group(2)), int(m.group(3))
        # Validación de rangos de mes y día
        if 1 <= mes <= 12 and 1 <= dia <= 31:
            resultado["valido"]      = True
            resultado["fecha"]       = f"{m.group(1)}-{m.group(2)}-{m.group(3)}"
            resultado["secuencial"]  = m.group(4)
    return resultado
```

```python
def validar_empleado(codigo: str) -> Dict:
    """
    Valida código de empleado.
    Formato: EMP-XXX-NNNN
    Patrón regex: ^EMP-([A-Z]{3})-([1-9]\d{3})$
      - [A-Z]{3}    → 3 letras mayúsculas (departamento)
      - [1-9]\d{3} → 4 dígitos donde el primero NO es 0
    El departamento además debe pertenecer a DEPARTAMENTOS_VALIDOS.
    """
    resultado = {"valido": False, "departamento": None, "numero": None}
    patron = r'^EMP-([A-Z]{3})-([1-9]\d{3})$'
    m = re.match(patron, codigo)
    if m:
        depto = m.group(1)
        if depto in DEPARTAMENTOS_VALIDOS:
            resultado["valido"]       = True
            resultado["departamento"] = depto
            resultado["numero"]       = m.group(2)
    return resultado
```

```python
def validar_factura(codigo: str) -> Dict:
    """
    Valida código de factura.
    Formato: FAC-S-NNNNNN
    Patrón regex: ^FAC-([A-E])-(\d{6})$
      - [A-E]  → una sola letra mayúscula entre A y E (serie)
      - \d{6} → exactamente 6 dígitos
    """
    resultado = {"valido": False, "serie": None, "numero": None}
    patron = r'^FAC-([A-E])-(\d{6})$'
    m = re.match(patron, codigo)
    if m:
        resultado["valido"] = True
        resultado["serie"]  = m.group(1)
        resultado["numero"] = m.group(2)
    return resultado
```

```python
print("PRUEBA DE FUNCIONES INDIVIDUALES")
print("=" * 50)

print("\n-- Productos --")
print(validar_producto("TEC-0001-MX"))
print(validar_producto("tec-0001-MX"))

print("\n-- Envíos --")
print(validar_envio("ENV-2024-03-15-001234"))
print(validar_envio("ENV-2024-13-15-001234"))

print("\n-- Empleados --")
print(validar_empleado("EMP-VEN-1234"))
print(validar_empleado("EMP-VEN-0123"))

print("\n-- Facturas --")
print(validar_factura("FAC-A-123456"))
print(validar_factura("FAC-F-123456"))
```

```python
def validar_codigo(codigo: str) -> Dict:
    """
    Detecta el tipo de código por su prefijo y llama al validador correspondiente.
    Prefijos reconocidos:
      ENV → envio
      EMP → empleado
      FAC → factura
      Tres letras mayúsculas → producto
    Cualquier otro prefijo → desconocido
    """
    resultado = {"codigo": codigo, "tipo": "desconocido", "valido": False, "detalles": {}}
    # Extraemos el primer segmento antes del primer guión
    prefijo = codigo.split("-")[0] if "-" in codigo else ""
    if prefijo == "ENV":
        r, tipo = validar_envio(codigo), "envio"
    elif prefijo == "EMP":
        r, tipo = validar_empleado(codigo), "empleado"
    elif prefijo == "FAC":
        r, tipo = validar_factura(codigo), "factura"
    elif re.match(r'^[A-Z]{3}$', prefijo):  # tres letras mayúsculas → producto
        r, tipo = validar_producto(codigo), "producto"
    else:
        return resultado
    resultado["tipo"]     = tipo
    resultado["valido"]   = r["valido"]
    resultado["detalles"] = {k: v for k, v in r.items() if k != "valido"}
    return resultado
```

```python
def mostrar_resultado(resultado: Dict) -> None:
    """Muestra el resultado de validación de forma legible."""
    estado = "✓" if resultado["valido"] else "✗"
    print(f"{estado} {resultado['codigo']:<30} | Tipo: {resultado['tipo']:<12}")
    if resultado["valido"] and resultado["detalles"]:
        detalles = ", ".join(f"{k}: {v}" for k, v in resultado["detalles"].items() if v)
        print(f"   └── {detalles}")

def mostrar_reporte(reporte: Dict) -> None:
    """Muestra el reporte de procesamiento por lotes."""
    print("=" * 60)
    print("                 REPORTE DE VALIDACIÓN")
    print("=" * 60)
    print(f"\nTotal procesados: {reporte['total']}")
    print(f"Válidos: {reporte['validos']} ({reporte['validos']/reporte['total']*100:.1f}%)")
    print(f"Inválidos: {reporte['invalidos']} ({reporte['invalidos']/reporte['total']*100:.1f}%)")
    print("\nDesglose por tipo:")
    print("-" * 40)
    for tipo, stats in reporte["por_tipo"].items():
        if stats["total"] > 0:
            tasa = stats["validos"] / stats["total"] * 100
            print(f"  {tipo.capitalize():<12}: {stats['validos']:>3}/{stats['total']:<3} ({tasa:.0f}% válidos)")
    print("\n" + "=" * 60)
```

```python
CODIGOS_PRUEBA = [
    # Productos
    "TEC-0001-MX", "ALI-9999-US", "ROB-1234-CA",
    "tec-0001-MX", "TEC-001-MX", "TECH-0001-MX",
    # Envíos
    "ENV-2024-03-15-001234", "ENV-2025-12-01-999999",
    "ENV-2019-03-15-001234", "ENV-2024-13-15-001234", "ENV-2024-03-32-001234",
    # Empleados
    "EMP-VEN-1234", "EMP-TEC-9999", "EMP-ADM-1000",
    "EMP-VEN-0123", "EMP-XXX-1234", "EMP-VEN-123",
    # Facturas
    "FAC-A-123456", "FAC-E-000001", "FAC-B-999999",
    "FAC-F-123456", "FAC-A-12345", "FAC-a-123456",
    # Desconocidos
    "XXX-1234", "RANDOM-CODE",
]

print("PRUEBA DE VALIDADOR UNIVERSAL")
print("=" * 50)
for codigo in CODIGOS_PRUEBA[:10]:
    mostrar_resultado(validar_codigo(codigo))
```

```python
def procesar_lote(codigos: List[str]) -> Dict:
    """
    Procesa múltiples códigos y genera estadísticas.
    Itera sobre cada código, llama a validar_codigo() y acumula
    los resultados en los contadores globales y por tipo.
    """
    resultado = {
        "total": 0, "validos": 0, "invalidos": 0,
        "por_tipo": {
            "producto":    {"total": 0, "validos": 0},
            "envio":       {"total": 0, "validos": 0},
            "empleado":    {"total": 0, "validos": 0},
            "factura":     {"total": 0, "validos": 0},
            "desconocido": {"total": 0, "validos": 0},
        },
        "detalle": []
    }
    for codigo in codigos:
        r = validar_codigo(codigo)
        resultado["total"] += 1
        if r["valido"]:
            resultado["validos"] += 1
        else:
            resultado["invalidos"] += 1
        tipo = r["tipo"]
        resultado["por_tipo"][tipo]["total"] += 1
        if r["valido"]:
            resultado["por_tipo"][tipo]["validos"] += 1
        resultado["detalle"].append(r)
    return resultado
```

```python
print("PRUEBA DE PROCESAMIENTO POR LOTES")
reporte = procesar_lote(CODIGOS_PRUEBA)
mostrar_reporte(reporte)
```

```python
def sugerir_correccion(codigo: str) -> str:
    """
    Intenta corregir el código convirtiendo a mayúsculas.
    Si el candidato corregido es válido, lo sugiere.
    """
    candidato = codigo.upper()
    r = validar_codigo(candidato)
    if r["valido"]:
        return f"¿Quisiste decir: {candidato}?"
    return "No se encontró una corrección automática."

def validar_fecha_real(anio: int, mes: int, dia: int) -> bool:
    """
    Usa datetime.date para verificar que la fecha exista realmente.
    Ejemplo: 2023-02-29 lanza ValueError → retorna False.
    """
    try:
        datetime.date(anio, mes, dia)
        return True
    except ValueError:
        return False

def exportar_resultados(reporte: Dict, archivo: str) -> None:
    """Guarda el detalle de validación en un archivo CSV."""
    with open(archivo, "w", newline="", encoding="utf-8") as f:
        campos = ["codigo", "tipo", "valido", "detalles"]
        writer = csv.DictWriter(f, fieldnames=campos)
        writer.writeheader()
        for item in reporte["detalle"]:
            writer.writerow({
                "codigo":   item["codigo"],
                "tipo":     item["tipo"],
                "valido":   item["valido"],
                "detalles": str(item["detalles"])
            })
    print(f"Resultados exportados a: {archivo}")
```

```python
print("Sugerencia de corrección:")
print(sugerir_correccion("tec-0001-MX"))
print(sugerir_correccion("fac-a-123456"))

print("\nValidación de fecha real:")
print(f"2024-02-29 (bisiesto): {validar_fecha_real(2024, 2, 29)}")
print(f"2023-02-29 (no bisiesto): {validar_fecha_real(2023, 2, 29)}")

exportar_resultados(reporte, "resultados_validacion.csv")
```
