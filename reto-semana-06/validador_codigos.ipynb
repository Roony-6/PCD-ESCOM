```python


def validar_envio(codigo: str) -> Dict:
    """Valida código de envío."""
    resultado = {"valido": False, "fecha": None, "secuencial": None}
    # Patrón Regex:
    # ^ENV-                 : Inicia con el prefijo estático ENV-
    # (202[0-9]|2030)       : Año de 2020 a 2030 (Grupo 1)
    # -(0[1-9]|1[0-2])      : Mes de 01 a 12 (Grupo 2)
    # -(0[1-9]|[12]\d|3[01]): Día de 01 a 31 (Grupo 3)
    # -(\d{6})$             : Exactamente 6 dígitos al final (Grupo 4)
    patron = r'^ENV-(202[0-9]|2030)-(0[1-9]|1[0-2])-(0[1-9]|[12]\d|3[01])-(\d{6})$'
    match = re.match(patron, codigo)
    
    if match:
        anio, mes, dia, secuencial = match.groups()
        # Validación de fecha real (aprovechando la lógica del Bonus)
        if validar_fecha_real(int(anio), int(mes), int(dia)):
            resultado.update({
                "valido": True,
                "fecha": f"{anio}-{mes}-{dia}",
                "secuencial": secuencial
            })
    return resultado


def validar_empleado(codigo: str) -> Dict:
    """Valida código de empleado."""
    resultado = {"valido": False, "departamento": None, "numero": None}
    
    # Patrón Regex:
    # ^EMP-         : Inicia con el prefijo estático EMP-
    # ([A-Z]{3})    : Exactamente 3 letras mayúsculas (Grupo 1 - Departamento)
    # -([1-9]\d{3})$: Guion seguido de 4 dígitos, el primero de 1-9 para evitar ceros iniciales (Grupo 2 - Número)
    patron = r'^EMP-([A-Z]{3})-([1-9]\d{3})$'
    match = re.match(patron, codigo)
    
    if match:
        depto = match.group(1)
        if depto in DEPARTAMENTOS_VALIDOS:
            resultado.update({
                "valido": True,
                "departamento": depto,
                "numero": match.group(2)
            })
    return resultado


def validar_factura(codigo: str) -> Dict:
    """Valida código de factura."""
    resultado = {"valido": False, "serie": None, "numero": None}
    
    # Patrón Regex:
    # ^FAC-   : Inicia con el prefijo estático FAC-
    # ([A-Z]) : Exactamente 1 letra mayúscula (Grupo 1 - Serie)
    # -(\d{6})$: Guion seguido de exactamente 6 dígitos (Grupo 2 - Número)
    patron = r'^FAC-([A-Z])-(\d{6})$'
    match = re.match(patron, codigo)
    
    if match:
        serie = match.group(1)
        if serie in SERIES_VALIDAS:
            resultado.update({
                "valido": True,
                "serie": serie,
                "numero": match.group(2)
            })
    return resultado
```
